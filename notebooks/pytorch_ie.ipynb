```python
from dataclasses import dataclass

from pytorch_ie.annotations import LabeledSpan
from pytorch_ie.auto import AutoPipeline
from pytorch_ie.core import AnnotationList, annotation_field
from pytorch_ie.documents import TextDocument

from typing import List
```

```python
import modelbit
mb = modelbit.login()
```

# 💪🏾 Building and Training

```python
@dataclass
class PersonDetectorDocument(TextDocument):
  entities: AnnotationList[LabeledSpan] = annotation_field(target="text")
```

```python
pipeline = AutoPipeline.from_pretrained("pie/example-ner-spanclf-conll03", device=-1, num_workers=0)
```

```python
def get_people(document: str) -> List[str]:
  doc = PersonDetectorDocument(document)
  pipeline(doc)
  results: List[str] = []
  for entity in doc.entities.predictions:
    if entity.label == "PER":
      results.append(str(entity))
  return results
```

```python
get_people("""
Sarthak Jain and Prathamesh Juvatkar founded Nanonets with the mission of leveraging AI to make workflow automation easier for businesses.
Seventy percent of organizations still have paper-dependent processes today, from parsing millions of contracts to manually matching  invoices to purchase orders.
Current legacy systems and processes require employees to review each document and manually enter data into CRMs, ERPs, or Excel files.
""")
```

# 🚀 Deployment

```python
mb.switch_branch("development")
```

```python
mb.deploy(get_people)
```
